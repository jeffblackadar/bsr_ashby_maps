```python
#https://www.geeksforgeeks.org/how-to-extract-images-from-pdf-in-python/
#https://pymupdf.readthedocs.io/en/latest/faq.html

# STEP 1
# import libraries
import fitz
import io
from PIL import Image

# STEP 2
# file path you want to extract images from
file = "/content/drive/MyDrive/gabii_legacy_gis_working/Copy of 8_foto muretti.pdf"

# open the file
pdf_file = fitz.open(file)
import cv2
# STEP 3
# iterate over PDF pages
doc = fitz.open(file)

for page in doc:  # iterate through the pages
    pix = page.get_pixmap()  # render page to an image
    pix.save("/content/drive/MyDrive/gabii_legacy_gis_working/8_foto muretti_page-%i.png" % page.number)  # store image as a PNG




```
